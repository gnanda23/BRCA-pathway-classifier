```python
import pandas as pd

# Path to VEP output (tab-delimited with # header)
vep_file = "../data/raw/brca_vep_output.txt"

# Read the file, skipping the header lines
with open(vep_file) as f:
    lines = [line for line in f if not line.startswith("##")]

# Extract header and data
header = lines[0].strip().lstrip("#").split("\t")
rows = [line.strip().split("\t") for line in lines[1:]]

# Create DataFrame
df_vep = pd.DataFrame(rows, columns=header)

# Preview
print(f"Loaded {df_vep.shape[0]} variants with {df_vep.shape[1]} features")
df_vep.head()
```
